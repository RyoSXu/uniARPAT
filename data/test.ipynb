```python
import csv

def get_max_cols(filename):
    max_cols = 0
    with open(filename, 'r', encoding='utf-8') as f:
        reader = csv.reader(f)
        for row in reader:
            # row 是一个列表，len(row) 即为该行的列数
            # 注意：index_col=0 意味着第一列是 ID，所以实际特征列数是 len(row) - 1
            max_cols = max(max_cols, len(row))
    return max_cols

# 测试你的原始文件
files_to_test = {
    "elements": "elements.csv",
    "position": "position.csv",
    "edos": "edos.csv",
    "phdos": "phdos.csv"
}

print("-" * 30)
for label, path in files_to_test.items():
    try:
        cols = get_max_cols(path)
        print(f"文件: {label:10} | 最大总列数 (含ID): {cols}")
        print(f"建议 {label} 的 names 长度应至少为: {cols}")
        print("-" * 30)
    except FileNotFoundError:
        print(f"错误: 找不到文件 {path}")
```
